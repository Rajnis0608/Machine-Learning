```python
from sklearn import datasets
```

```python
boston = datasets.load_boston()
X = boston.data 
Y = boston.target

import pandas as pd
boston_df = pd.DataFrame(X)
boston_df.columns = boston.feature_names

#spliting data into train and test data
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X,Y)

#implemention algo
from sklearn.linear_model import LinearRegression
algo_1 = LinearRegression()
algo_1.fit(X_train,Y_train)

#Getting predicted output for X_test
Y_pred = algo_1.predict(X_test)

#Comparing actual output with predicted
import matplotlib.pyplot as plt
plt.scatter(Y_test,Y_pred)
plt.plot(Y_test,Y_test,label="Line to test variation",color="red")
plt.xlabel("Actual Output")
plt.ylabel("Predicted Output")
plt.legend()
plt.show()
```
