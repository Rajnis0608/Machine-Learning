```python
# Importing libraries
import numpy as np
import pandas as pd
```

```python
# Importing dataset
data = pd.read_csv('iris.data',header=None)
data.columns = ['sepal_length','sepal_width','petal_length','petal_width','label']
data
```

# Handling missing values


```python
# Checking NaN 
data.isna()
```

```python
#cgacking NaN
data.isnull()
```

```python
# count of NaN
data.isna().sum()
```

```python
# dropping rows of label with NaN
data.dropna(subset=['label'], inplace = True)
```

```python
# sustituting NaN
data = data.fillna(3.7)
```

```python
# Label Encoder
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
encoded_data = data.copy(deep = True)
le.fit(encoded_data.iloc[:,4])
encoded_data.iloc[:,4]= le.transform(encoded_data.iloc[:,4])
encoded_data
```
